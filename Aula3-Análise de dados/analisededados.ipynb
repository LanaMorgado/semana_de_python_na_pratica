```python
import pandas as pd

# Carrega os dados do arquivo Excel
dados = pd.read_excel("vendas.xlsx")

# Verifica as primeiras e últimas linhas
print(dados.head())
print(dados.tail())
```

```python
# Quantidade de linhas e colunas
print(dados.shape)

# Informações sobre as colunas
print(dados.info())

# Gerando estatísticas descritivas
print(dados.describe())

```

```python
# Obtendo os valores únicos de uma coluna
print(dados["loja"].unique())
```

```python
# Contagem de valores por loja
print(dados["loja"].value_counts())

```

```python
# Agrupando os dados por loja e somando os preços
print(dados.groupby("loja")["preco"].sum())
```

```python
import plotly.express as px
grafico = px.histogram(dados,x="loja", y="preco", color="forma_pagamento", text_auto=True, title="Faturamento por loja")

grafico.show()
```
